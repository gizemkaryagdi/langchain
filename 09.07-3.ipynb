```python
from langchain.embeddings import OpenAIEmbeddings 
```

```python
embeddings=OpenAIEmbeddings() #openai yerleştirmelerini başlatmaya çalışarak bizim için kullanılabilecek nesneye sahip olacağız
```

```python
text="Hey Buddy"
```

```python
text_embedding=embeddings.embed_query(text) #son olarak bu metni elimizdeki embeds nesnesine bu şekilde aktardığımda buradaki özel metni sayısal temsiline dönüştürmesi gerekir.
```

```python
print(f"Our embedding is length{text_embedding}")
```
